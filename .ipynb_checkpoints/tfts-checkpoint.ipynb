### Zaman Serilerinde Yaygın Örüntüler

Zaman serileri farklı şekil ve boyutlarda gelebilir. Ancak birkaç örüntü çok yaygındır.
Özellikle bunları gördüğümüzde tanımkta fayda var.

* 1- Zaman serilerinin belirli bir yönde hareket ettiği **TREND**. (bkz: Moore Yasası)
* 2- Desenler öngörülebilir aralıklarla tekrarlandığında görülen **MEVSİMSELLİK**.
* 3- Bazı zaman serileri hem trende hem de mevsimselliğe aynı anda sahip olabilir.
* 4- Muhtemelen hiç bir zaman tahmin edilemeyen, **GÜRÜLTÜ** (white noise) olarak adlandırılan rastgele değerler kümesi de karşımıza çıkabilir.
* 5- **OTOKORELASYON**: Trend yok ve mevsimsellik yok. Sıçramalar rastgele zamanlarda görünür. Bundan sonra ne zaman olacağını veya ne kadar güçlü olacağını tahmin edemiyorsunuz. Ama açık ki, tüm seri rastgele de değil. Sıçramalar arasında çok belirleyici bir düşüş/azalma vardır.

### UYGULAMA - Kendi Zaman Serimizi Üretelim

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
# zaman serisini görmemizi sağlayan genel-yardımcı fonksiyon
def plot_series(time, series, format="-", start=0, end=None, label=None):
    plt.plot(time[start:end], series[start:end], format, label=label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize=14)
    plt.grid(True)
```

```python
# Trend üreten fonksiyon
def trend(time, slope=0):
    return slope * time
```

```python
# yukarı yönlü ilerleyen bir trend oluşturalım
time = np.arange(4 * 365 + 1)
baseline = 10
series = trend(time, 0.1)

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
# Mevsimsellik için gerekli fonksiyonlar
def seasonal_pattern(season_time):
    """Keyfi bir örüntü kullanıyoruz. İsterseniz değiştirebilirsiniz"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Her periyotta aynı örüntüyü tekrar ediyoruz"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

```python
# Mevsimsellik içeren bir seri oluşturalım
baseline = 10
amplitude = 40
series = seasonality(time, period=365, amplitude=amplitude)

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
# Şimdi de hem mevsimsellik hem de trend içeren bir zaman serisi oluşturalım
slope = 0.05
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude)

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
# Gerçekte zaman serileri böyle pürüzsüz bir akışa sahip değildir. 
# Genellikle rassal gürültü içerirler.
```

```python
# Gürültü üreten fonksiyon
def white_noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
# Biraz gürültü üretelim
noise_level = 5
noise = white_noise(time, noise_level, seed=42)

plt.figure(figsize=(10, 6))
plot_series(time, noise)
plt.show()
```

```python
# Zaman serimize bu gürültüyü uyarlayalım
series += noise

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

```python
# Son olarak otokorelasyon ekleyeceğiz
```

```python
def autocorrelation(time, amplitude, seed=None):
    rnd = np.random.RandomState(seed)
    φ1 = 0.5
    φ2 = -0.1
    ar = rnd.randn(len(time) + 50)
    ar[:50] = 100
    for step in range(50, len(time) + 50):
        ar[step] += φ1 * ar[step - 50]
        ar[step] += φ2 * ar[step - 33]
    return ar[50:] * amplitude
```

```python
def autocorrelation(time, amplitude, seed=None):
    rnd = np.random.RandomState(seed)
    φ = 0.8
    ar = rnd.randn(len(time) + 1)
    for step in range(1, len(time) + 1):
        ar[step] += φ * ar[step - 1]
    return ar[1:] * amplitude
```

```python
# Otokorelasyon üretelim
series = autocorrelation(time, 10, seed=42)
plot_series(time[:200], series[:200])
plt.show()
```

```python
# Otokorelasyon ve trend birlikte
series = autocorrelation(time, 10, seed=42) + trend(time, 2)
plot_series(time[:200], series[:200])
plt.show()
```

```python
# Trend, mevsimsellik ve otokorelasyon
series = autocorrelation(time, 10, seed=42) + seasonality(time, period=50, amplitude=150) + trend(time, 2)
plot_series(time[:200], series[:200])
plt.show()
```

```python
# 200. günde radikal bir değişikliğin olduğu durumu simule edelim
series = autocorrelation(time, 10, seed=42) + seasonality(time, period=50, amplitude=150) + trend(time, 2)
series2 = autocorrelation(time, 5, seed=42) + seasonality(time, period=50, amplitude=2) + trend(time, -1) + 550
series[200:] = series2[200:]
#series += noise(time, 30)
plot_series(time[:300], series[:300])
plt.show()
```
